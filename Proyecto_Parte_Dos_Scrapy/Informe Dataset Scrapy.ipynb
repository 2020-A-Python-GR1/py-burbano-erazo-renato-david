## Contexto
El siguiente dataset corresponde datos de la página web Metacritic. Es un sitio web que recopila reseñas de álbumes de música, videojuegos, películas, programas de televisión, DVD y libros. En nuestro estudio se ha tomado datos de los dos últimos años (2019, 2020) sobre las puntuacines a Videojuegos ya sea por los críticos como por los usuarios, puede encontrarse en el siguiente enlace: https://www.metacritic.com/browse/games/score/metascore/year/all/filtered?year_selected=2019&distribution=&sort=desc&view=condensed, primero se ha hecho un escano de la información de la página web con Scrapy para recolectar los títulos lanzados en esos dos años con sus respectivas puntuaciones

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

df_metacritic_2019 = pd.read_csv('metacritic/data_scrapy_2019.csv')
df_metacritic_2020 = pd.read_csv('metacritic/data_scrapy_2020.csv')
```

#### 1. ¿Cuáles son los juegos 2019 con mejor calificación de Metacritic?

```python
df_meta_score = df_metacritic_2019.sort_values('MetaScore',ascending=False)
df_meta_score = df_meta_score.drop_duplicates(['Titulos'], keep='first')

df_meta_score = df_meta_score.head(10)
df_meta_score = df_meta_score.loc[:,['Titulos','MetaScore']]

y_pos = np.arange(len(df_meta_score))

plt.barh(y_pos, df_meta_score.MetaScore, align='center', alpha=0.5)

plt.yticks(y_pos, df_meta_score.Titulos)
plt.xlabel('Calificación Metacritic')

plt.title('Los juegos mejor calificados por Metacritic 2019')
plt.show()
```

#### 9. ¿Cuáles son los juegos de PC con peores calificaciones de usuarios en el 2019?

```python
df_pc = df_metacritic_2019['Plataforma'] == 'PC'
df_pc = df_metacritic_2019[df_pc]

df_pc = df_pc.sort_values('UserScore',ascending=True)
df_pc = df_pc.head(10)
df_pc = df_pc.loc[:,['Titulos','UserScore']]
titulos = df_pc['Titulos']

x = np.arange(len(df_pc.index))

width = 0.35
 
fig, ax = plt.subplots()
 
rects1 = ax.bar(x - width/2, df_pc.UserScore, width, label='UserScore')

ax.set_title('Peor Calificación por Usuarios de PC 2019')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
fig.tight_layout()

plt.show()
i = 0
for titulo in titulos:
    print(str(i) + ' - ' + titulo)
    i = i+1

```

#### 17. ¿Cuál es el promedio de calificaciones hecha por Metacritic y los Usuario en 2019?

```python
df_mean_19 = df_metacritic_2019.loc[:,['MetaScore','UserScore']].mean()
df_mean_19.MetaScore = df_mean_19.MetaScore / 10
df_mean_19

fig, ax = plt.subplots()
ax.bar(df_mean_19.index,df_mean_19.values)
ax.set_title('Promedio de calificación de Metacritic y de Usuarios en 2019')
ax.grid(True)

plt.show()
```

#### 18. ¿Cuál es la media de calificaciones hecha por Metacritic y los Usuario en 2020?

```python
df_mode_20 = df_metacritic_2020.loc[:,['MetaScore','UserScore']].median()
df_mode_20.MetaScore = df_mode_20.MetaScore / 10
df_mode_20

fig, ax = plt.subplots()
ax.bar(df_mode_20.index,df_mode_20.values)
ax.set_title('Media de calificación de Metacritic y de Usuarios en 2020')
ax.grid(True)

plt.show()
```

#### 19. ¿Cuál es la moda de calificaciones hecha por Metacritic y los Usuario en 2020?
